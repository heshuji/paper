```python
import numpy
import pylab as pl
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import calinski_harabasz_score
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans

plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体,这个依赖是不能识别中文的,要自己设置下哦
plt.rcParams["axes.unicode_minus"]=False #这个句子可以解决图像中的“-”负号这个东西显示乱码的一个小问题
# %matplotlib inline
df=pd.read_excel('./合并表格.xlsx') # 导入我们处理后的表格哦,这个是拼接好的
```

```python
data_KMeans=df[df["类型"]=="高钾"] # 提出来类型等于高钾的数据
data_KMeans=data_KMeans.iloc[:,1:15]

calinski_harabaz_list_KMeans=[]
for i in range(2,6):
    #构建,并训练,模型
    kmeans = KMeans(n_clusters = i,random_state=213).fit(data_KMeans)

    y_pred = KMeans(n_clusters = i,random_state=231).fit_predict(data_KMeans)
    score = calinski_harabasz_score(data_KMeans,kmeans.labels_)
    calinski_harabaz_list_KMeans.append(score)

data_MBM=df[df["类型"]=="高钾"] # 提出来类型等于高钾的数据
data_MBM=data_MBM.iloc[:,1:15]

calinski_harabaz_list_MBM=[]
for i in range(2,6):
    #构建,并训练,模型
    kmeans = MiniBatchKMeans(n_clusters = i,random_state=213).fit(data_MBM)

    y_pred = MiniBatchKMeans(n_clusters = i,random_state=231).fit_predict(data_MBM)
    score = calinski_harabasz_score(data_MBM,kmeans.labels_)
    calinski_harabaz_list_MBM.append(score)

```

```python
fig, ax1 = plt.subplots(1,1,figsize=(16,9),dpi=300)

list_k=[2,3,4,5]
line1, = ax1.plot(list_k,calinski_harabaz_list_KMeans,label='KMeans')
line2, = ax1.plot(list_k,calinski_harabaz_list_MBM,label='MiniBatchKMeans')
ax1.set_xlabel("聚类数",fontsize = 12)
ax1.set_ylabel("calinski_harabaz指数",fontsize = 12)
plt.legend()
```
