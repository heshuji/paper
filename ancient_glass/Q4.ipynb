```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif']=['SimHei']  # 用于显示中文
plt.rcParams['axes.unicode_minus'] = False  # 用于显示中文

df=pd.read_excel("./合并表格.xlsx")
df_hk=df[df["类型"]=="高钾"]
df1=df_hk.iloc[:,1:15]
pearson_hk=df1.corr(method="pearson")
plt.figure(dpi=300, figsize=(16, 9))
ax=sns.heatmap(pearson_hk,cmap=plt.get_cmap('Blues_r'), annot=True,fmt=".2f")
```

```python
df_db=df[df["类型"]=="铅钡"]
df2=df_db.iloc[:,1:15]
pearson_db=df2.corr(method="pearson")
plt.figure(dpi=300, figsize=(16, 9))
ax=sns.heatmap(pearson_db,cmap=plt.get_cmap('Blues_r'), annot=True,fmt=".2f")
```

```python
cha=pearson_db-pearson_hk
```

```python
plt.figure(dpi=300, figsize=(16, 9))
ax=sns.heatmap(cha,cmap=plt.get_cmap('Blues_r'), annot=True,fmt=".2f")



```
