```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline

# plt.rcParams['font.sans-serif'] = ['SimHei'] # 步骤一（换字体.,不然中文显示不了
plt.rcParams['axes.unicode_minus'] = False   # 步骤二（负号显示问题）
dian_t=pd.read_excel('./合并表格.xlsx')
```

```python

fig,axes=plt.subplots(dpi=600,figsize=(8,9))
dian_t1=dian_t[(dian_t["Types"]=="High potassium") & (dian_t["Surface weathering"]=="weathering")]
dian_t1.to_excel("高钾_风化.xlsx",index=False)
dian=dian_t1.iloc[:,1:15]
axes.boxplot(dian,patch_artist=True,widths=0.8) #描点上色
axes.yaxis.grid(True) #在y轴上添加网格线
# # 加刻度名称
xtick=np.arange(dian.shape[1])+1
a=dian.columns
plt.setp(axes, xticks=xtick,
         xticklabels=a)
plt.xticks(rotation=70)# xlable 倾斜70度
plt.title("High Potassium Weathering Boxplot")
plt.show() #展示
```

```python
fig,axes=plt.subplots(dpi=600,figsize=(8,9))
dian_t1=dian_t[(dian_t["Types"]=="High potassium") & (dian_t["Surface weathering"]=="No weathering")]
dian_t1.to_excel("高钾_无风化.xlsx",index=False)
dian=dian_t1.iloc[:,1:15]
axes.boxplot(dian,patch_artist=True) #描点上色
axes.yaxis.grid(True) #在y轴上添加网格线
# # 加刻度名称
xtick=np.arange(dian.shape[1])+1
a=dian.columns
print(a)
plt.setp(axes, xticks=xtick,
         xticklabels=a)
plt.xticks(rotation=70)# xlable 倾斜70度
plt.title("High potassium non-weathering box diagram")
# dian.describe().to_excel("高钾无风化_描述性统计.xlsx")
plt.show() #展示
```
